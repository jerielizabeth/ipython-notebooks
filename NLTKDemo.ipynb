## Part-of-Speech Tagging

```python
NY = wordlists.words('NY1848.txt')
```

```python
ny_tagged = nltk.pos_tag(NY)
```

```python
print (ny_tagged)
```

```python
tag_fd = nltk.FreqDist(tag for (word, tag) in ny_tagged)
```

```python
tag_fd.most_common()
```

## Using POS Tagging in Entity-Extraction

```python
ny_chunked = nltk.ne_chunk(ny_tagged, binary=True)
```

```python
print(ny_chunked)
```

```python
entity_names = []
```

```python
for c in ny_chunked:
    if type(c) is nltk.Tree:
        if c.label() == 'NE':
            entity_names.append(' '.join(i[0] for i in c.leaves()))
```

```python
print (set(entity_names))
```

## Overview of Sentiment Analysis

Sentiment Analysis or opinion mining is the "computational study of opinions, sentiments, 
subjectivity, evaluations, attitudes, appraisal, affects, views, emotions, etc., expressed in text" (http://www.cs.uic.edu/~liub/FBS/Sentiment-Analysis-tutorial-AAAI-2011.pdf). Used for analyzing social media, reviews, automated recommendations, etc. A variation on this was used in the Facebook emotions study (http://liwc.net/howliwcworks.php).

Sentiments are positive, negative, and neutral. The text can be analyzed at the document, sentence, and entity or feature level. For analysis, you need at least 1. target entity; 2. an aspect of that entity; 3. an opinion holder; 4. time of opinion expression; and 5. sentiment value of the opinion from the opinion holder on the feature of the entity as the time of expression. (This is the definition of an opinion used for sentiment analysis.)

Combination of determining sentiment value, extracting entity and aspects, and combining that information. It is a hard problem.


```python
## code from http://streamhacker.com/2010/05/10/text-classification-sentiment-analysis-naive-bayes-classifier/
import nltk.classify.util
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import movie_reviews
```

```python
def word_feats(words):
    return dict([(word,True) for word in words])
```

```python
negids = movie_reviews.fileids('neg')
posids = movie_reviews.fileids('pos')
```

```python
negfeats = [(word_feats(movie_reviews.words(fileids=[f])), 'neg') for f in negids]
posfeats = [(word_feats(movie_reviews.words(fileids=[f])), 'pos') for f in posids]
```

```python
negcutoff = int(len(negfeats)*3/4)
poscutoff = int(len(posfeats)*3/4)
```

```python
trainfeats = negfeats[:negcutoff] + posfeats[:poscutoff]
testfeats = negfeats[negcutoff:] + posfeats[poscutoff:]
print ("train on {} instances, test on {} instances".format(len(trainfeats), len(testfeats)))
```

```python
classifier = NaiveBayesClassifier.train(trainfeats)
print ('accuracy:', nltk.classify.util.accuracy(classifier, testfeats))
```

```python
classifier.show_most_informative_features()
```
